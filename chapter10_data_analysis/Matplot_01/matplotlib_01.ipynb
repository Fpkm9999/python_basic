맷플롯립 라이브러리는 그래프를 그릴 때 가장 많이 사용하는 파이썬 라이브러리 중 하나
- 맷플롯립을 활용하면 다양한 유형의 그래프를 간편하게 그릴 수 있음.
- 시각화할 때 데이터로 리스트나 딕셔너리 자료형도 가능하지만 데이터프레임을 이용하면
- 새로운 그래프에 적용하는 것이 더 쉽기 때문에 대부분 데이터프레임 자료형을 이용.

matplotlib는 MATLAB (과학 및 공학 연산을 위한 소프트웨어)의 시각화 기능을 모델링해서 만들어짐.

공식 홈페이지 : https://matplotlib.org/


### 01. 시각화 옵션

```python
# 그래프를 표현하다 보면 그래프 제목, X 축 및 Y 축 이름 등의 그래프에 대한 추가 옵션을 지정할 때가 많음.

# 1) 제목

# 그래프 제목은 title() 메서드를 사용하여 표현.
# 제목은 기본적으로 영어를 사용하는데, 한글을 사용하려면 별도의 설정을 해야 함.
```

```python
# 제목 지정
import matplotlib.pyplot as plt

plt.title('Line graph')  # <-- 얘를 한글로 쓰면 안된다.
xdata = [2, 4, 6, 8]  # <-- x,y 의 데이터를 리스트로 저장함
ydata = [1, 3, 5, 7]
plt.plot(xdata, ydata)
plt.show()  # <-- 표를 print 해줌
```

```python
# 2) 범례

# 그래프 범례는 두 개 이상의 데이터를 표현할 때 사용.
# 범례 추가는 plot() 을 활용하여 표현하는데, plot()의 label 속성값에 원하는 문자열을 작성하고
# 그래프를 그리기 전에 legend() 메서드를 실행. <- 안하면 안됨

# 범례 지정
plt.title('Legend')
data1 = [2, 4, 6, 8]
data2 = [8, 6, 4, 2]
plt.plot(data1, label='asc')
plt.plot(data2, label='desc')
plt.legend()  # <- 범례를 나오게함
plt.show()  # 해당 그래프에 대한 설명 : 라벨
```

```python
# 3) 색상
# 그래프 색상은 plot() 메서드에 color 속성을 추가하여 표현.

# 색상 지정
xdata = [2, 4, 6, 8]
ydata = [1, 3, 5, 7]
plt.plot(xdata, ydata, color='red')
plt.show()
```

```python
dir(plt.plot())
```

```python
# 4) x 축 및 y 축 이름
# x 축 이름은 xlabel()를 사용하고, y 축 이름은 ylabel() 을 사용하여 표현.

# x축 및 y 축 이름 지정
xdata = [2, 4, 6, 8]
ydata = [1, 3, 5, 7]
plt.plot(xdata, ydata)
plt.xlabel('X value')
plt.ylabel('Y value')
plt.show()
```

```python
# 5) 그래프 선 모양
# 그래프 선 모양은 plot()의 linestyle 속성을 이용하여 표현.
# linestyle 속성으로 실선은 '-', 파선은 '--', 점쇄선 '-.', 점선은 ':' 기호로 지정

# 그래프 선 모양 지정
data1 = [2, 4, 6, 8]
data2 = [8, 6, 4, 2]

plt.plot(data1, color='r', label='dashed', linestyle='--')
plt.plot(data2, color='g', label='dotted', linestyle=':')
plt.legend()
plt.show()
```

```python
# 6) 그림 범위 지정
# plot() 를 사용하여 그래프를 표현하다 보면 몇몇 점들은 그림의 범위 경계선에 있어서 잘 보이지 않을 때가 있음.
# xlim() 메서드와 ylim() 메서드를 사용하여 그림의 범위를 수동으로 지정할 수 있음.
# xlim() 은 x 축의 최솟값과 최대값을 지정하고, ylim()은 y축의 최솟값과 최대값을 지정.
# plot()의 marker 속성을 지정하면 선에 점 모양을 다양하게 변경할 수 있음
```

```python
# 그림 범위 지정
plt.title('X, Y range')
xdata = [10, 20, 30, 40, ]
ydata = [1, 3, 5, 7]
plt.plot(xdata,ydata, color='b',linestyle='--', marker='o', markersize=10)
plt.show()
```

```python
# 그림 범위 지정 -> 축소
plt.title('X, Y range')
xdata = [10, 20, 30, 40, ]
ydata = [1, 3, 5, 7]
plt.plot(xdata,ydata, color='b',linestyle='--', marker='o', markersize=10)
plt.xlim(0,50)
plt.ylim(-5,15)
plt.show()
```

```python
# 그림 범위 지정 -> 확대
plt.title('X, Y range')
xdata = [10, 20, 30, 40, ]
ydata = [1, 3, 5, 7]
plt.plot(xdata,ydata, color='b',linestyle='--', marker='o', markersize=10)
plt.xlim(10,30)
plt.ylim(1,4)
plt.show()
```

```python
# 7) 내장 시각화 옵션
# 판다스를 이용하면 시각화 라이브러리를 임포트 하지 않아도 내장 그래프 도구를 이용하여 기본적인 시각화를 할 수 있음.
# plot()의 kind 속성을 지정하면 다양한 종류의 그래프를 표현할 수 있음.

# 그래프 속성 값
# 옵션 : 종류
# line : 선 그래프
# bar : 막대 그래프 - 수직
# barh : 막대 그래프 - 수평
# his : 히스토그램 그래프
# box : 박스 그래프
# kde : 커널 밀도 그래프
# area : 면적 그래프
# pie : 원형 그래프
# scatter : 산점도 그래프
# hexbin : 고밀도 산점도 그래프

# -> 자주 쓰는것 위주로만 봄
import pandas as pd

my_score = [[60, 90, 95], [80, 75, 100], [65, 85, 90], [85, 70, 90], [95, 90, 85], [75, 85, 90], [85, 80, 75]]
subject = ['kor', 'math', 'eng']
df = pd.DataFrame(my_score, columns=subject)
df
```

```python
df.plot(kind='line')
plt.show()
```
