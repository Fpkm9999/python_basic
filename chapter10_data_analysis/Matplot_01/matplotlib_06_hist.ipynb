```python
# 1. 공공데이터 포털에서 '국민건강보험공단_건강검진정보' 검색해서 csv 파일을 다운받으세요.
# 2024년 3월 기준으로 2022년 데이터가 업로드 되어있음.

# 2. 파일을 읽은 후 데이터프레임으로 변환하시오.

# 3. 남성의 키데이터를 가지고 히스토그램을 만들것.

# 4. 남성과 여성이 키 데이터를 하나의 히스토그램으로 만들것

import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('./input/국민건강보험공단_건강검진정보_20221231.CSV', encoding='cp949')
data.head(20)
```

```python
data = data.rename(columns={'성별 ': 'gender', '신장(5cm단위)': 'height'})
data.rename(columns={'성별': 'gender', '신장(5cm단위)': 'height'}, inplace=True)

# man_data = data.loc[data.gender == 1.0, ['gender', 'height']]
# man_data.head()
man_data = data.loc[data.gender == 1.0, ['gender', 'height']]
# data에서 여성(data.gender==2.0)의 성별, 키 데이터만 가져옴.
woman_data = data.loc[data.gender == 2.0, ['gender', 'height']]
```

```python
# 1) 히스토그램 그리기

# data에서 남성(data.gender==1.0)의 성별, 키 데이터만 가져옴
man_data = data.loc[data.gender == 1.0, ['gender', 'height']]
man_data.head()
```

```python
data['gender'].value_counts()
```

```python
man_data.info()
```

```python
plt.figure(figsize=(10, 6))  # 그래프의 크기를 지정
plt.hist(man_data['height'], bins=10, alpha=0.5, label='Man')
plt.hist(woman_data['height'], bins=10, alpha=0.5, label='Woman')  # alpht는 투명도(0:투명,1:불투명)

plt.title('2020 Health Screenings man & woman Height Group Histogram')
plt.xlabel('height')
plt.ylabel('frequency')
plt.xlim(120, 200)
plt.legend()
plt.grid()
plt.show()
```

```python
# 5. 연령대별 흡연자 숫자를 막대그래프로 출력
data.value_counts('흡연상태')
# 해당 수검자의 흡연 상태 여부 1(피우지 않는다), 2(이전에 피웠으나 끊었다), 3(현재도 피우고 있다)
```

```python
# smoke_data = data.rename(columns={'흡연상태': 'smoke_state'})

# smoke_data2 = smoke_data.loc[data.smoke_state == 3.0]
dt1 = data.loc[data.흡연상태 == 3.0, ['흡연상태', '연령대코드(5세 단위)']]
dt1
```

```python
df2 = dt1.groupby('연령대코드(5세 단위)').count()
df2  # <-- 의료보험이 40세 이상부터 해줘서 그 이하인 20 30은 안나옴
```

```python
df2.index = ['40~44', '45~49', '50~54', '55~59', '60~64', '65~69',
             '70~74', '75~79', '80~84', '85+']
```

```python
plt.figure(figsize=(15, 4))
plt.bar(df2.index, df2['흡연상태'], color='g')
plt.title('Smokers by age group Bar Graph')
# plt.xlabel('빈도')
# plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # '맑은 고딕'으로 지정
matplotlib.rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(15, 4))
plt.bar(df2.index, df2['흡연상태'])
plt.title('연령대별 흡연자 수')
plt.xlabel('나이')
plt.ylabel('빈도')
plt.grid()
plt.show()
```
