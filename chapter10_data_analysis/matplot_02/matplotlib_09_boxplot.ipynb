### 06. 상자 수염 Box Plot 그래프

'데이터 분포를 시각화'하여 탐색하는 방법으로 상자수염 그래프가 있음.
사람의 나이가 500살, 키가 5미터 등은 표현은 가능하지만 현실적으로 불가능한 값들.
> 이를 이상치, 극단치라고 하는데, 이상치는 효과적인 데이터 분석을 위해 제거해야 함.

- 이상치를 제거하기 위해 값의 정상 범위를 정하는 방법이 있음.
- 나이의 경우 0살에서 120살 까지를 정상 범위로 정한다면 121살은 이상치로 보는 것.
- 또 다른 방법으로 통계적인 기법을 사용하는 것인데,
- 정규 분포에서 상하위 0.2% 밖의 데이터를 극단적인 값으로 가정하여 제거하는 것

```python
# 예시
import numpy as np
import matplotlib.pyplot as plt

# 랜덤한 데이터 생성
np.random.seed(0)
data1 = np.random.randn(100)  # 첫 번째 데이터
np.random.seed(1)
data2 = np.random.randn(100)  # 두 번째 데이터

# 상자수염 그래프 그리기
plt.boxplot([data1, data2])
plt.xticks([1, 2], ['Data1', 'Data2'])
plt.ylabel('Value')
plt.title('Box Plot Example')
plt.grid(True)
plt.show()
# 위 코드에서 data1과 data2는 두 개의 데이터셋입니다. plt.boxplot() 함수를 사용하여 
# 상자수염 그래프를 그립니다. 그래프의 x축 레이블은 ['Data1', 'Data2']로 설정되었습니다.
```
