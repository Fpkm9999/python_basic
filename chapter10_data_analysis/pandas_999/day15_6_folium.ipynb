```python
#### 04. 지도에 원형 마커 표시
```

```python
import pandas as pd
import folium
```

```python
# 대학교 리스트를 데이터 프레임 변환
df = pd.read_excel('./input/서울지역 대학교 위치.xlsx')
df.head()
```

```python
# 서울 지도 만들기
seoul_map = folium.Map(location=[37.55, 126.98], zoom_start=12, tiles='Cartodb Positron')
```

```python
# 대학교 위치 정보를 CircleMarker 로 표시
# Marker() 대신 CircleMarker() 사용
# 원형 마커의 크기, 색상, 투명도, 등 설정 가능
for name, lat, lng in zip(df['name'], df['위도'], df['경도']):
    folium.CircleMarker((lat, lng), popup=name,
                        radius = 10,  # 원의 반지름
                        color = 'brown',  # 원의 둘레 색상
                        fill = True,
                        Fill_color='coral',  # 원 채우기 색
                        fill_opacity=0.7,  # 투명도
                        ).add_to(seoul_map)
seoul_map
```

```python
# 지도를 html 파일로 저장하기
seoul_map.save('./output/seoul_colleges2.html')
```
