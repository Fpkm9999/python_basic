````python
import requests
from bs4 import BeautifulSoup as bs
import pprint

url: str = 'https://music.bugs.co.kr/chart/'
response = requests.get(url)
# print(response.text) 
# print(response) # <Response [200]> 
soup = bs(response.text, 'html.parser')
# print(response.status_code)
# print(soup)
# print(soup.prettify())
datas = soup.select('table > tbody > tr')
print(type(datas))  # <class 'bs4.element.ResultSet'>

# print(datas)
for data in datas:
    # print(type(data))   # <class 'bs4.element.Tag'>
    print(data.select_one('td').text)
    # print(data.text.strip())
````

`datas` 변수에 `.text`나 `.strip()` 메소드가 사용되지 않는 이유는 `datas`가 `list` 타입의 객체이기 때문입니다. `soup.select` 메서드는 CSS 선택자에 해당하는 모든 요소를 찾아서 리스트 형태로 반환합니다. 각각의 요소는 `bs4.element.Tag` 타입의 객체로, 이 객체들은 HTML 또는 XML의 개별 태그를 나타냅니다. `list` 객체 자체에는 `.text`나 `.strip()` 메서드가 없습니다. 이 메서드들은 `bs4.element.Tag` 객체에 적용될 수 있으며, 태그 내의 텍스트를 추출할 때 사용됩니다.

따라서, `datas` 리스트에 포함된 각 태그 객체에 대해 개별적으로 `.text`를 호출해야 합니다. 예를 들어, `datas` 리스트의 첫 번째 요소에 대해 `.text`를 호출하려면 다음과 같이 작성할 수 있습니다:

```python
if datas:  # datas가 비어있지 않은 경우에만 실행
    print(datas[0].text.strip())  # 첫 번째 요소의 텍스트를 추출하고 공백을 제거
```

전체 리스트에 대해 반복문을 사용하여 각 태그의 텍스트를 추출할 수도 있습니다:

```python
for data in datas:
    print(data.text.strip())  # 각 요소의 텍스트를 추출하고 공백을 제거
```

이 방식을 사용하면 `datas` 리스트에 있는 각 `bs4.element.Tag` 객체의 텍스트 내용을 성공적으로 출력할 수 있습니다. `bs4.element.Tag` 객체의 `.text` 속성은 해당 태그와 그 자식 태그에 포함된 모든 텍스트를 연결한 문자열을 반환하며, `.strip()` 메서드는 이 문자열의 시작과 끝에 있는 공백을 제거합니다.
