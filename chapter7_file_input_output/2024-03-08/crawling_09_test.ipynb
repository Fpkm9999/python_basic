```python
import pprint
import requests
from bs4 import BeautifulSoup as bs
import os.path
import csv
import json
from requests import Response

# 1. 멜론 HTML 파싱하기 
url = 'https://www.melon.com/chart/'

def requests_get(url: str) -> Response:
    request_headers: dict = {
        'User-Agent': ('Mozilla/5.0 (Windows NT 12.0; Win64; x64) AppleWebKit/537.36 '
                       '(KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36'),
        'Referer': '',
    }
    return requests.get(url, headers=request_headers)

response = requests_get(url)    # 함수화
soup = bs(response.text, 'html.parser')

# print(soup.prettify())

# print(f'제목 / 아티스트 / 앨범') 가져오기
datas = soup.select('table tbody tr')
# pprint.pprint(datas)

for idx,item in enumerate(datas):
    print(item.select_one(".rank01").text.strip())
```
