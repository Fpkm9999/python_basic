```python
# 연습 1
import matplotlib.pyplot as plt

# 데이터 생성
time_values = [1, 2, 3, 4, 5]  # 시간
temperature_values = [20, 22, 25, 23, 21]  # 온도

# 선 그래프 그리기
plt.plot(time_values, temperature_values)
plt.xlabel("time")
plt.ylabel("temp")
plt.title("Temperature changes over time")
plt.grid(True)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # 맑은 고딕으로 지정
matplotlib.rcParams['axes.unicode_minus'] = False
x = np.arange(0, 5, 1)
y1 = x
y2 = x + 1
y3 = x + 2
y4 = x + 3
```

```python
print(x)
print(y1)
print(y2)
print(y3)
print(y4)
```

```python
plt.plot(x, y1, '>-r', x, y2, 's-g', x, y3, 'd:b', x, y4, '-.Xc')
plt.legend(['데이터1', '데이터2', '데이터3', '데이터4'], loc='best')
plt.xlabel('X 축')
plt.ylabel('Y 축')
plt.title('그래프 제목')
plt.grid(True)
plt.show()
```
