### 4. 산점도 그래프

서로 다른 두 연속성 변수 사이의 관계를 나타내며 연속성 변수의 상관성을 확인할 떄 산점도 그래프를 사용.
예를 들어 나이와 소득에 대한 상호 관련성 파악 등에 유용하게 사용되는데, 두 변수 간의 상관관계를 개략적으로 파악할 수 있기 때문,
산점도에 표시되는 각 점들은 자료의 관측값을 의미하고, 각 점의 위치는 관측값이 가지는 x축, y축 변수의 값으로 결정

참고 : https://newsjel.ly/archives/newsjelly/15193

```python
# 공공데이터포털 https://www.data.go.kr/ 에서 다운로드한 2020년 건강검진 일부 데이터 health_screenings_2020.xlsx를
# 이용하여 산점도 그래프를 표현
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# data = pd.read_excel('./input/health_screenings_2020.xlsx')
# data.head()
```

```python
data=pd.read_csv('./input/health_screenings_2020.csv')
data
```

```python
data.info()
```

```python
df1 = data.iloc[:100000]  # 데이터가 많으니 1만건만
df1.info()
```

```python
height_data = df1['height']  # 키 데이터만 가져와 height_data에 저장
weight_data = df1['weight']  # 몸무게 데이터만 가져와 weight_data 에 저장
height_data.head()
```

```python
weight_data.head()
```

```python
plt.figure(figsize=(10, 4))  # 그래프 크기를 지정
plt.scatter(height_data, weight_data)  # x 축 데이터에 height_data, y 축 데이터에 weight_data 지정
plt.title('2020 Health Screenings Scatter Graph')
plt.xlabel('height')
plt.ylabel('weight')
plt.show()
```

```python
df1
```

```python
# 2) 그룹 산점도 그래프 그리기
# 누적 산점도 그래프를 그리는 것은 기존 산점도 그래프위에 산점도 그래프를 하나 더 그리는 방식.

# HDL, LDL, cholesterol 데이터를 가져와 각각의 변수에 저장
HDL_data = df1['HDL']
LDL_data = df1['LDL']
cholesterol_data = df1['cholesterol']

plt.figure(figsize=(10, 6))  # 그래프의 크기를 지정

# scatter() 에 x 축 데이터와 y 축 데이터를 지정하고 점 색상과 점 테두리 색상을 지정.
plt.scatter(cholesterol_data, LDL_data, color='r', edgecolor='w', label='Cholesterol*LDL')
plt.scatter(HDL_data, cholesterol_data, color='g', edgecolor='w', label='HDL*Cholesterol')
plt.scatter(HDL_data, LDL_data, color='b', edgecolor='w', label='LDL*HDL')

plt.title('2020 Health Screenings Group Scatter Graph')
# xlim(), ylim() 로 x 축 및 y축의 최솟값과 최댓값을 지정.
plt.xlim(-50, 500)
plt.ylim(-50, 500)
plt.legend()
plt.show()
# 산점도를 써서 서로의 상관관계여부를 확인할 수 있다.
```
