### 공공 데이터 포탈에서 다운받은 '소상공인시장진흥공단_상가(상권)정보_경북_202312'을 사용.

 1. 필요한 정보만 데이터 프레임으로 변경
 2. 경산, 표준산업분류명: 커피전문점 지도에 표시

```python
import pandas as pd
import matplotlib
import folium
```

```python
#1. 필요한 데이터 정보만 데이터 프레임으로 변경
df = pd.read_csv('./input/소상공인시장진흥공단_상가(상권)정보_경북_202312.csv')
df.head()
```

```python
df.info()
```

```python
df.columns = df.columns.map(str)
df
```

```python
df.info()
```

```python
df['시군구명'].values
```

```python
#  1. 필요한 정보만 데이터 프레임으로 변경
df2 = df[['표준산업분류명', '경도', '위도', '시군구명', '상호명']]
df2
```

```python
df3 = df2[(df2['시군구명'] == '경산시') & (df2['표준산업분류명'] == '커피 전문점')]
df3.head()
df3.tail()
```

```python
df3.info()
```

```python
df3
```

```python
import folium.map

# 경산 지도 객체 생성
gyeongsan_map = folium.Map(location=[35.9542891, 128.833745], zoom_star=13, tiles='Cartodb Positron')
gyeongsan_map
```

```python
# popup : 마우스 클릭시 나옴 : tooltip : 마우스 호버 시 나옴
for name, lat, lng in zip(df3['상호명'], df3['위도'], df3['경도']):
    tooltip = folium.Tooltip(name, permanent=True)
    folium.Marker(location=[lat, lng], tooltip=tooltip).add_to(gyeongsan_map)
gyeongsan_map
```
