#### 지도 위 마커 표시하기

```python
# 서울 시내 주요 대학교의 위치 데이터를 데이터 프레임으로 변환하고, folium 지도에 위치 표시

import pandas as pd
import folium
```

```python
# 대학교 리스트를 데이터 프레임으로 변환
df = pd.read_excel('./input/서울지역 대학교 위치.xlsx')
df.head()
```

```python
# 서울 지도 만들기
seoul_map = folium.Map(location=[37.55, 126.98], zoom_start=12)
```

```python
# 대학교 위치 정보를 Marker 로 표시
for name, lat, lng in zip(df['name'], df['위도'], df['경도']):
    folium.Marker([lat, lng], popup=name).add_to(seoul_map)
seoul_map
```

```python
# 지도를 html 파일로 저장
seoul_map.save('./output/seoul2.html')
```
