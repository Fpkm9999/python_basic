```python
# 3. 북대구 영업소의 2023년 12월 일자별 총 교통량의 변화를 선그래프로 나타내시오.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib
matplotlib.rcParams['font.family'] ='sans-serif'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
tcs = pd.read_csv('./input/TCS_B3_04_03_238959.csv',encoding='cp949')
tcs.head()
```

```python
sns.lineplot(data= tcs,x ='집계일자',y='총교통량')
plt.show()
```

```python
tcs.info()
```

```python
df1 = tcs[tcs['영업소명'] == '북대구']
df1
```

```python
df1['date2'] = pd.to_datetime(df1['집계일자'])
df1.info()
```

```python
df1[['집계일자','date2']].head()
```

```python
df2 = df1[(df1['date2'] >= '2023-12-01') & (df1['date2'] <= '2023-12-30')]

df2.head(100)

# 혹은
# df3 = df1[df1['집계일자'].astype(str).str.contains('2023-12')]
# df3.head()  # 북대구 12월자 데이터
```

```python
df4 = df2.groupby('집계일자', as_index=False).agg(n=('총교통량', 'sum'))
df4
```

```python
plt.figure(figsize=(16,10))
sns.lineplot(data=df4,x='집계일자',y='n')
plt.show()
```
