# 03. 지도에 마커 표시하기

- 서울 시내 주요 대학교의 위치 데이터를 데이터 프레임으로 반환하고, folium 지도에 위치를 표시.

```python
import pandas as pd
import folium
```

```python
# 해각교 리스트를 데이터프레임 변환
df = pd.read_excel('./input/서울지역 대학교 위치.xlsx')
df.head()  # df는 데이터 프레임 객체임.
```

```python
df.columns
```

```python
df.values
```

```python
# 서울 지도 만들기
seoul_map = folium.Map(location=[37.55, 126.98], zoom_start=12)
seoul_map
```

```python
# 대학교 위치 정보를 Marker 로 표시
for name, lat, lng in zip(df['name'], df['위도'], df['경도']):
    # folium.Marker([lat,lng],popup=name).add_to(seoul_map)
    folium.Marker([lat, lng], popup=name).add_to(seoul_map)
    # 위도, 경도
# folium.Maker(위도,경도, popup=None).add_to(지도_map객체)
seoul_map
```
