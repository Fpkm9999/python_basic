```python
# 1번 

# 패키지 임포트
import matplotlib
import pandas as pd  # pandas(판다스) 패키지 임포트
import matplotlib.pyplot as plt  # Matplotlib(맷플롯립) 패키지의 pyplot 모듈을 plt로 임포트
import seaborn as sns  # Seaborn(씨본) 패키지 임포트
import warnings

# 경고창 무시
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('./input/아파트(매매)_실거래가_20240304160735.csv', encoding='cp949')
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
# 2) 구별 거래 건수와 거래 금액 합계를 막대 그래프로 출력
df['구'] = df['시군구'].str.split(' ').str[1]
df['구']
```

```python
df['거래금액(만원)'] = df['거래금액(만원)'].str.replace(',', '').astype(int)
df['거래금액(만원)']
```

```python
df.info()
```

```python
data_daegu_result = df.groupby('구', as_index=False).agg(order_total=('거래금액(만원)', 'sum'))
data_gu_total = df.groupby('구', as_index=False).agg(order_count=('구', 'count'))
data_gu_total
data_daegu_result['order_count'] = data_gu_total['order_count']
data_daegu_result
```

```python
data_daegu_result.info()
data_daegu_result['구'] = data_daegu_result['구'].astype(str)
data_daegu_result
```

```python
data_daegu_result.describe()
```

```python
# 2) 구별 거래 건수와 거래 금액 합계를 막대 그래프로 출력 답

# 구별 금액 합계
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # '맑은고딕'으로 변경
matplotlib.rcParams['axes.unicode_minus'] = False
# 막대 그래프 만들기
plt.figure(figsize=(10, 10))
sns.barplot(data=data_daegu_result, x='구', y='order_total')
plt.show()
```

```python
# 구 별 거래 건수
plt.figure(figsize=(10, 10))
sns.barplot(data=data_daegu_result, x='구', y='order_count')
plt.show()
```

```python
# 3) 대구 전체의 구별 거래 건수와 거래 금액을 파이 그래프로 출력

# 구별 거래 금액
labels = ['군위군', '남구', '달서구', '달성군', '동구', '북구', '서구', '수성구', '중구']
# 
plt.pie(data_daegu_result['order_total'], labels=labels, autopct='%.1f%%')
plt.show()
```

```python
# 구별 거래 건수
labels = ['군위군', '남구', '달서구', '달성군', '동구', '북구', '서구', '수성구', '중구']
# 
plt.pie(data_daegu_result['order_count'], labels=labels, autopct='%.1f%%')
plt.show()
```

```python
# 5) 수성구 동별로 거래건수와 거래 금액의 합계를 막대 그래프로 출력
df3 = df[df['구'] == '수성구']
df4 = df3.groupby('동', as_index=False).agg(order_total=('거래금액(만원)', 'sum'))
susung_total = df3.groupby('동', as_index=False).agg(order_count=('동', 'count'))
df4['order_count'] = susung_total['order_count']
df4
```

```python
# 수성구 동 별로 거래 금액의 합계
index = np.arange(len(df4))
plt.figure(figsize=(15, 4))
plt.bar(index, df4['order_total'])
plt.grid()
plt.show()
```

```python
# 6) 수성구 전체의 거래 건수와 평균 거래금액의 월별 변화를 선 그래프로 출력

df3 = df[df['구'] == '수성구']
df5 = df3.groupby('계약년월', as_index=False).agg(order_count=('NO', 'count'))
df5
susung_avg = df3.groupby('계약년월', as_index=False).agg(order_avg=('거래금액(만원)', 'mean'))
# susung_avg
df5['order_avg'] = susung_avg['order_avg']

df5
```

```python
#수성구 전체의 거래 건수 의 월별 변화를 선그래프로 출력
plt.figure(figsize=(15, 4))
plt.plot(df5['계약년월'], df5['order_count'])
plt.show()
```

```python
# 수성구의 평균 거래금액의 월별 변화를 선그래프로 출력
plt.figure(figsize=(15, 4))
plt.plot(df5['계약년월'], df5['order_avg'])
plt.show()
```
