```python
# 패키지 임포트
import matplotlib
import pandas as pd  # pandas(판다스) 패키지 임포트
import matplotlib.pyplot as plt  # Matplotlib(맷플롯립) 패키지의 pyplot 모듈을 plt로 임포트
import seaborn as sns  # Seaborn(씨본) 패키지 임포트
import warnings

matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # '맑은고딕'으로 변경
matplotlib.rcParams['axes.unicode_minus'] = False
# 경고창 무시
warnings.filterwarnings('ignore')
```

```python
# 압축을 푼 후 대구 관련 파df일을 선택해서 작업을 진행하시오. (24년 3월 11일 기준으로 113,640건의 데이터가 있습니다.)
df = pd.read_csv('./input/소상공인시장진흥공단_상가(상권)정보_대구_202312.csv')
df.head(20)
```

```python
df['상권업종소분류코드']
```

```python
df.describe()
```

```python
df.info()
```

```python
df1 = df[df['상권업종소분류코드'] == 'I21201']
df1
```

```python
df1.info()
```

```python
# 1) '상권업종소분류코드'를 이용해서 카페의 데이터만 추출하여 엑셀로 저장하시오. (파일명: 대구_카페.xlsx)

df1.to_excel('./output/대구_카페.xlsx')
```

```python
# 2) 구별 카페수를 막대그래프로 출력

df1
```

```python
df2 = df1.groupby('시군구명', as_index=False).agg(cafe_count=('상가업소번호', 'count'))
df2
```

```python
# 2) 구별 카페수를 막대그래프로 출력 <-- 답


plt.figure(figsize=(15, 4))
plt.bar(df2['시군구명'], df2['cafe_count'], color='g')
plt.grid()
plt.show()
```

```python
# 3) 대구시 전체의 구별 카페수를 파이그래프로 출력

plt.figure(figsize=(15, 4))
plt.pie(df2['cafe_count'], labels=df2['시군구명'], autopct='%.1f%%')
plt.grid()
plt.show()
```

```python
# 4) 수성구는 카페수를 법정동명을 기준으로 동별로 막대그래프로 출력

df2 = df1[df1['시군구명'] == '수성구']
df2  # <-- 카페 & 수성구
df3 = df2.groupby('법정동명', as_index=False).agg(cafe_count=('법정동명', 'count'))
df3
```

```python
df3.info()
```

```python
df3.describe()
```

```python
# 4) 답
plt.figure(figsize=(15, 10))
plt.xticks(rotation=45)
sns.barplot(data=df3, y=df3['cafe_count'], x=df3['법정동명'])
plt.show()
```

```python
# 5) 수성구전체의 카페수를 법정동면 기준으로 파이 그래프로 출력

df4 = df2.groupby('법정동명', as_index=False).agg(cafe_count=('법정동명', 'count'))
df4
```

```python
plt.figure(figsize=(17, 10))
plt.pie(df4['cafe_count'], labels=df4['법정동명'], autopct='%.1f%%')
plt.show()
```
