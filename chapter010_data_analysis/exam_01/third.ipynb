```python
# 패키지 임포트
import matplotlib
import pandas as pd  # pandas(판다스) 패키지 임포트
import matplotlib.pyplot as plt  # Matplotlib(맷플롯립) 패키지의 pyplot 모듈을 plt로 임포트
import warnings
import folium

matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # '맑은고딕'으로 변경
matplotlib.rcParams['axes.unicode_minus'] = False
# 경고창 무시
warnings.filterwarnings('ignore')
```

```python
# 1) 이디야 커피 체인점의 데이터만 추출해서 엑셀로 저장하시오. (파일명 : 대구_카페_이디야.xlsx)

df = pd.read_csv('./input/소상공인시장진흥공단_상가(상권)정보_대구_202312.csv')
df.head(20)
```

```python
df2 = df[df['상호명'].astype(str).str.contains('이디야')]
df2
```

```python
df2.describe()
```

```python
df2['상호명'].value_counts().sum()
```

```python
df2.to_excel('./output/대구_카페_이디야.xlsx')
```

```python
# 2) folium라이브러리를 사용해 지도에 마커를 표시하시오. 마커에 마우스 이벤트 없이도 상호명이 나오도록 하시오.
susung_map = folium.Map(location=[35.816257, 128.525558], zoom_start=12)
susung_map
```

```python
df2
```

```python
for name, lat, lng in zip(df2['상호명'], df2['위도'], df2['경도']):
    tooltip = folium.Tooltip(name, permanent=True)
    folium.Marker([lat, lng], tooltip=tooltip).add_to(susung_map)
susung_map
```

```python
# 3) 해당 지도 객체를 html 파일로 저장하시오. (파일명 : 대구_카페_이디야_지도.html)
susung_map.save('./output/대구_카페_이디야_지도.html')
```
