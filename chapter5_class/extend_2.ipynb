```python
class Person:
    population: int = 0  # 클래스 변수

    def __init__(self, name: str) -> None:
        self.name = name
        print(f'{self.name} is born')
        Person.population += 1

    def __del__(self):
        print(f'{self.name} is dead')
        Person.population -= 1

    @classmethod
    def get_population(cls):
        return cls.population


man = Person('james')
woman = Person('emily')
print('전체 인구수: {}명'.format(Person.get_population()))

del man
print('전체 인구수: {}명'.format(Person.get_population()))
del woman
print('전체 인구수: {}명'.format(Person.get_population()))
```

```python
# 2번
class Shop:
    total = 0  # 클래스 변수
    menu_list = [{'떡볶이': 3000}, {'순대': 3000}, {'튀김': 500}, {'김밥': 2000}]

    # 클래스 변수(딕셔너리)
    @classmethod
    def sales(cls, product: str, amount: int) -> None:
        for item in cls.menu_list:
            if product in item:  # 제품 이름이 딕셔너리의 키와 일치하는지 확인 # 제품 이름이 딕셔너리의 키에 해당하는지 확인
                cls.total += item[product] * amount  # 해당 제품의 가격을 총액에 추가

    @classmethod
    def sales_info(cls):  # 클래스 메소드로 변경
        print(f'매출: {cls.total}')  # 클래스 변수에 접근


# 제품 판매
Shop.sales('떡볶이', 1)  # 떡볶이를 1개 판매
Shop.sales('김밥', 2)  # 김밥을 2개 판매
Shop.sales('튀김', 5)  # 튀김을 5개 판매

# 판매 정보 출력
Shop.sales_info()
```

```python
# 3번
class Car:  # 슈퍼 클래스
    max_oil = 50  # 최대 주유량

    def __init__(self, oil):  # 생성자
        self.oil = oil

    def add_oil(self, oil):
        if oil <= 0:  # 0 이하의 주유는 진행하지 않음
            return
        self.oil += oil  # 0 이상이라면 주유
        if self.oil > Car.max_oil:  # 주유 후 최대 주유량 초과 상태이면
            self.oil = Car.max_oil  # 현재 주유량을 최대 주유량으로 설정

    def car_info(self):
        print('현재 주유상태: {}'.format(self.oil))


class Hybrid(Car):  # 서브 클래스
    max_oil = 30  # 최대 주유량

    def __init__(self, oil: int, charge_oil: int):
        super().__init__(oil)
        self.charge_oil = charge_oil

    def add_oil(self, oil):
        if oil <= 0:  # 0 이하의 주유는 진행하지 않음
            return
        self.oil += oil  # 0 이상이라면 주유
        if self.oil > Hybrid.max_oil:  # 주유 후 최대 주유량 초과 상태이면
            self.oil = Hybrid.max_oil  # 현재 주유량을 최대 주유량으로 설정

    def charge(self, charge_oil):
        self.charge_oil += charge_oil

    def hybrid_info(self):
        print('현재 주유상태: {}'.format(self.oil))
        print('현재 충전상태: {}'.format(self.charge_oil))


car = Hybrid(0, 0)
car.add_oil(100)
car.charge(50)
car.hybrid_info()
# 현재 주유상태: 50 # 현재 충전상태: 30
```

```python
class Car:
    max_oil: int = 50  # 최대 주유량

    def __init__(self, oil: int) -> None:
        self.oil = oil

    def add_oil(self, oil):
        if oil <= 0:  # 0 이하의 주유는 진행하지 않음
            return
        self.oil += oil
        if self.oil > Car.max_oil:  # 주유 후 최대 주유량 초과 상태이면
            self.oil = Car.max_oil  # 현재 주유량을 최대 주유량으로 설정

    def car_info(self):
        print('현재 주유상태: {}'.format(self.oil))


class Hybrid(Car):
    max_battery: int = 30

    def __init__(self, oil: int, battery: int) -> None:
        super().__init__(oil)
        self.battery = battery

    def charge(self, amount: int) -> None:
        if amount <= 0:  # 0 이하의 주유는 진행하지 않음
            return
        self.battery += amount
        if self.battery > Hybrid.max_battery:  # 주유 후 최대 주유량 초과 상태이면
            self.battery = Hybrid.max_battery  # 현재 주유량을 최대 주유량으로 설정

    def hybrid_info(self):
        print(f'# 현재 주유상태:{Hybrid.max_oil} # 현재 충전상태:{self.battery}')


car = Hybrid(0, 0)
car.add_oil(100)
car.charge(50)
car.hybrid_info()  # 현재 주유상태: 50 # 현재 충전상태: 30
```
